```python
import FinanceDataReader as fdr

sse = fdr.DataReader('224060')
sse = sse['Close']
sse_before = sse.shift(1)

yield_df = 100 * ((sse / sse_before) - 1)
```

```python
import matplotlib.pyplot as plt

sse.plot()
sse_before.plot()
```

```python
import numpy as np

m = yield_df.sum() / len(yield_df)
variance = np.var(yield_df)
std = np.std(yield_df)

print(f"""
평균: {m}
분산: {variance}
표준편차: {std}
VaR: {-2.33*std}
""")
```

```python
var = -2.33 * std
maximum_risk_value_df = sse * (1 + 0.01 * var)
```
